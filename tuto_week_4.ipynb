```python
import pandas as pd
import kagglehub
import numpy as np


# Download latest version
path = kagglehub.dataset_download("grandmaster07/restaurant-tips-dataset-analysis")

print("Path to dataset files:", path)
```

```python
import os
os.listdir(path)
df= pd.read_csv(f"{path}/tip.csv")
```

```python
target='tip'
features =[col for col in df.columns if col != target]
print("Target column:", target)
print("Feature columns:", features)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(df['total_bill'],bins=20,kde=True)
plt.title("distribution of Total Bills")
plt.show()

sns.scatterplot(x="total_bill",y="tip",hue="sex",data=df)
plt.title("Total Bill vs Tip by Gender")
plt.show()

sns.barplot(x='day',y='tip',data=df,estimator=np.mean,ci=None)
plt.title("Average Tip by Day")
plt.show()
```
